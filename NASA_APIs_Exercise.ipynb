```python
# Send a request to the NASA API"

import requests
import pandas as pd
# API setup
base_url = "https://api.nasa.gov/neo/rest/v1/feed"
api_key = "DEMO_KEY"

start_date = "2025-04-01"
end_date = "2025-04-04"


# Parameters for the GET request
params = {
    "start_date" : start_date,
    "end_date" : end_date,
    "api_key" : api_key

}
# Sending the GET request
response = requests.get(base_url, params=params)

# Check the status code to ensure the request was successful
response.status_code

# Parsing the JSON Response
if response.status_code == 200:
    data = response.json()
    # print(data)
    all_asteroids = []

# Loop through each date and its list of asteroids
    for date in data["near_earth_objects"]:
        for asteroid in data["near_earth_objects"][date]:
            asteroid_info = {
                "Asteroid ID": asteroid["id"],
                "Asteroid name": asteroid["name"],
                "Minimal estimated diameter (km)": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
                "Absolute_magnitude": asteroid["absolute_magnitude_h"],
                "Relative_velocity (km/s)": float(asteroid["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"])
            }
            all_asteroids.append(asteroid_info)
else:
    print("API Error:", response.status_code, response.text)

# --- 4. Create DataFrame ---
df = pd.DataFrame(all_asteroids)

# --- 5. Export to CSV ---
df.to_csv("asteroid_data.csv", index=False)

# --- 6. Preview the DataFrame ---
df.head()
# print (response.text)
# print(response.url)
```
